## Linear Regression

```python
# TESTING CELL
# This cell is using one random state for testing pipeline and model
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression


train_r2 = np.zeros(len(random_states))
val_r2 = np.zeros(len(random_states))
test_r2 = np.zeros(len(random_states))
state = np.zeros(len(random_states))

# Loop through pipeline with multiple random states.
for i in range(len(random_states)):  
    state[i] = i + 1
    # 60% training, 20% validation, 20% split test
    X_train, X_other, Y_train, Y_other = train_test_split(x, y, train_size = 0.6, 
                                                          random_state=random_states[i])
    
    print('training set:', X_train.shape, Y_train.shape)
    print(X_other.shape, Y_other.shape)
    X_val, X_test, Y_val, Y_test = train_test_split(X_other, Y_other, train_size = 0.5, 
                                                    random_state=random_states[i])
    
    print('validation set:', X_val.shape, Y_val.shape)
    print('test set:', X_test.shape, Y_test.shape)
    
    # Encoders for preprocessing
    preprocessor = ColumnTransformer(
    transformers=[
        #('ord', OrdinalEncoder(categories = ordinal_catagories), ordinal_features),
        ('onehot', OneHotEncoder(sparse=False, handle_unknown='ignore'), onehot_features),
        ('min_max', MinMaxScaler(), min_max_features),
        ('standard_scl', StandardScaler(), standard_features)])

    prep = Pipeline(steps=[('preprocessor', preprocessor)])  
                                                       

    X_train_prep = prep.fit_transform(X_train)
    X_val_prep = prep.transform(X_val)
    X_test_prep = prep.transform(X_test)

    print(X_train.shape)
    print(X_train_prep.shape)
    print(X_train_prep)
    print(X_val.shape)
    print(X_val_prep.shape)
    print(X_val_prep)
    print(X_test.shape)
    print(X_test_prep.shape)
    print(X_test_prep)
    
    # Initialize the Linear Regression model
    linreg = LinearRegression(fit_intercept=True)
    # linreg.fit(X_train_prep[:,:1], Y_train)
    linreg.fit(X_train_prep, Y_train)
    print('intercept:',linreg.intercept_)
    print('theta:',linreg.coef_)
    
    # R2 Score metric
    train_r2[i] = r2_score(Y_train, linreg.predict(X_train_prep))
    val_r2[i] = r2_score(Y_val, linreg.predict(X_val_prep))
    test_r2[i] = r2_score(Y_test, linreg.predict(X_test_prep))
    print('train R2:',train_r2[i])
    print('val R2:',val_r2[i])
    print('test R2:',test_r2[i])
    
```

```python
# Calculate mean and standard deviation from R2 scores
print('train R2 mean', np.around(np.mean(train_r2), 5))
print('validation R2 mean', np.around(np.mean(val_r2),5))
print('test R2 mean', np.around(np.mean(test_r2),5))
print('train R2 standard deviation', np.std(train_r2))
print('validation R2 standard deviation', np.std(val_r2))
print('test R2 standard deviation', np.std(test_r2))
print('test set accuracy', np.around(np.mean(test_r2),5), '+/-', np.around(np.std(test_r2),5))
```

```python
# Calculate standard deviation from baseline R2 score
print('train R2 standard deviation', np.std(train_r2))
print('validation R2 standard deviation', np.std(val_r2))
print('test R2 standard deviation', np.std(test_r2))
```

## Lasso Regularization

```python
# TESTING CELL USING LASSO REGULARIZATION
# This cell is using one random state for testing pipeline and model

from  sklearn.linear_model import Lasso

alpha = np.logspace(-7,0,29)
#alpha = np.logspace(-10,0,51) # this value for logspace was in class example
thetas = []
models = []

train_r2_lasso = np.zeros(len(alpha))
val_r2_lasso = np.zeros(len(alpha))
test_r2_lasso = np.zeros(len(alpha))
train_std_lasso = np.zeros(len(alpha))
val_std_lasso = np.zeros(len(alpha))
test_std_lasso = np.zeros(len(alpha))

alpha_num = np.zeros(len(alpha))

for l in range(len(alpha)):
    # Train, validation, and test R2 scores per random state 
    train_r2 = np.zeros(len(random_states))
    val_r2 = np.zeros(len(random_states))
    test_r2 = np.zeros(len(random_states))
    
    # Keep track of random state position
    state = np.zeros(len(random_states))
    
    # Keep track of the interation of max depth
    alpha_num[l] = l + 1
    
    for i in range(len(random_states)): 
        
        # Keep track of the random state interation
        state[i] = i + 1
        
        # 60% training, 20% validation, 20% split test
        X_train, X_other, Y_train, Y_other = train_test_split(x, y, train_size = 0.6,
                                                              random_state=random_states[i])
        print('training set:', X_train.shape, Y_train.shape)
        print(X_other.shape, Y_other.shape)
        X_val, X_test, Y_val, Y_test = train_test_split(X_other, Y_other, train_size = 0.5, 
                                                        random_state=random_states[i])
        print('validation set:', X_val.shape, Y_val.shape)
        print('test set:', X_test.shape, Y_test.shape)
    
        # Encoders for preprocessing
        preprocessor = ColumnTransformer(
        transformers=[
            #('ord', OrdinalEncoder(categories = ordinal_catagories), ordinal_features),
            ('onehot', OneHotEncoder(sparse=False, handle_unknown='ignore'), onehot_features),
            ('min_max', MinMaxScaler(), min_max_features),
            ('standard_scl', StandardScaler(), standard_features)])

        prep = Pipeline(steps=[('preprocessor', preprocessor)]) 
                                                       

        X_train_prep = prep.fit_transform(X_train)
        X_val_prep = prep.transform(X_val)
        X_test_prep = prep.transform(X_test)

        print(X_train.shape)
        print(X_train_prep.shape)
        print(X_train_prep)
        print(X_val.shape)
        print(X_val_prep.shape)
        print(X_val_prep)
        print(X_test.shape)
        print(X_test_prep.shape)
        print(X_test_prep)
    
   
        # Initialize lasso linear regression model
        lasso_reg = Lasso(alpha=alpha[l],max_iter=100000000)
        
        # Fit lasso to training set
        lasso_reg.fit(X_train_prep, Y_train)
        thetas.append(lasso_reg.coef_)
        models.append(lasso_reg)
        
        # Predict lasso for training, validation and test sets
        y_new_train = lasso_reg.predict(X_train_prep)
        y_new_val = lasso_reg.predict(X_val_prep)
        y_new_test = lasso_reg.predict(X_test_prep)

        # R2 Score metrics
        train_r2[i] = r2_score(Y_train, y_new_train)
        val_r2[i] = r2_score(Y_val, y_new_val)
        test_r2[i] = r2_score(Y_test, y_new_test)
        print('train R2:',train_r2[i])
        print('val R2:',val_r2[i])
        print('test R2:',test_r2[i])
        
    train_r2_lasso[l] = np.mean(train_r2)
    val_r2_lasso[l] = np.mean(val_r2)
    test_r2_lasso[l] = np.mean(test_r2)
    train_std_lasso[l] = np.std(train_r2)
    val_std_lasso[l] = np.std(val_r2)
    test_std_lasso[l] = np.std(test_r2)
    
```

## Random Forest Regression

```python
# Random Forest Regression pipeline using the same random states as other algorithms.
# This pipeline was designed to run each maximum depth value through the same interations of
#the random state value for comparitive results.

from sklearn.ensemble import RandomForestRegressor

# List of tree depths
max_depth = [1, 2, 3, 5, 10, 20, 50]

train_r2_rfor = np.zeros(len(max_depth))
val_r2_rfor = np.zeros(len(max_depth))
test_r2_rfor = np.zeros(len(max_depth))
train_std_rfor = np.zeros(len(max_depth))
val_std_rfor = np.zeros(len(max_depth))
test_std_rfor = np.zeros(len(max_depth))

depth_num = np.zeros(len(max_depth))


# Loop through pipeline with multiple tree depths.
for l in range(len(max_depth)):  
    # Train, validation, and test R2 scores per random state 
    train_r2 = np.zeros(len(random_states))
    val_r2 = np.zeros(len(random_states))
    test_r2 = np.zeros(len(random_states))
    state = np.zeros(len(random_states))
    
    # Keep track of the interation of max depth
    depth_num[l] = l + 1
    
    # Loop through random states within each max depth interation
    for i in range(len(random_states)): 
        
        # Keep track of the random state interation
        state[i] = i + 1
       
        # 60% training, 20% validation, 20% split test
        X_train, X_other, Y_train, Y_other = train_test_split(x, y, train_size = 0.6, 
                                                          random_state=random_states[i])
        print('training set:', X_train.shape, Y_train.shape)
        print(X_other.shape, Y_other.shape)
        X_val, X_test, Y_val, Y_test = train_test_split(X_other, Y_other, train_size = 0.5, 
                                                    random_state=random_states[i])
        print('validation set:', X_val.shape, Y_val.shape)
        print('test set:', X_test.shape, Y_test.shape)
    
        # Encoders for preprocessing
        preprocessor = ColumnTransformer(
        transformers=[
            #('ord', OrdinalEncoder(categories = ordinal_catagories), ordinal_features),
            ('onehot', OneHotEncoder(sparse=False, handle_unknown='ignore'), onehot_features),
            ('min_max', MinMaxScaler(), min_max_features),
            ('standard_scl', StandardScaler(), standard_features)])

        prep = Pipeline(steps=[('preprocessor', preprocessor)])  
                                                       

        X_train_prep = prep.fit_transform(X_train)
        X_val_prep = prep.transform(X_val)
        X_test_prep = prep.transform(X_test)

        print(X_train.shape)
        print(X_train_prep.shape)
        print(X_train_prep)
        print(X_val.shape)
        print(X_val_prep.shape)
        print(X_val_prep)
        print(X_test.shape)
        print(X_test_prep.shape)
        print(X_test_prep)
    
        # Initialize Random Forest Regressor
        rand_for = RandomForestRegressor(n_estimators=100,max_depth=max_depth[l])
    
        # Fit Random Forest Regressor to training set
        rand_for.fit(X_train_prep, Y_train)
    
        # Predict Random Forest Regressor for training, validation and test sets
        y_new_train = rand_for.predict(X_train_prep)
        y_new_val = rand_for.predict(X_val_prep)
        y_new_test = rand_for.predict(X_test_prep)
    
        # R2 Score metric
        train_r2[i] = r2_score(Y_train, y_new_train)
        val_r2[i] = r2_score(Y_val, y_new_val)
        test_r2[i] = r2_score(Y_test, y_new_test)
        print('train R2:',train_r2[i])
        print('val R2:',val_r2[i])
        print('test R2:',test_r2[i])
    
    train_r2_rfor[l] = np.mean(train_r2)
    val_r2_rfor[l] = np.mean(val_r2)
    test_r2_rfor[l] = np.mean(test_r2)
    train_std_rfor[l] = np.std(train_r2)
    val_std_rfor[l] = np.std(val_r2)
    test_std_rfor[l] = np.std(test_r2)
    
    
```

```python
# R2 Means for training, validation and test sets of R2 Random Forrest Regression
print('training R2 means:', train_r2_rfor)
print('validation R2 means:', val_r2_rfor)
print('testing R2 means:', test_r2_rfor)
```

```python
# Standard Deviations for training, validation and test sets of R2 Random Forrest Regression
print('training R2 standard deviations:', train_std_rfor)
print('validation R2 standard deviations:', val_std_rfor)
print('testing R2 standard deviations:', test_std_rfor)
print(state)
```
