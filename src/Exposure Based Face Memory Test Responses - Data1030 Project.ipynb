```python
# Validation of preprocessing methods for multiple random states.

for random_state in random_states:   
    # 60% training, 20% validation, 20% split test
    X_train, X_other, Y_train, Y_other = train_test_split(x, y, train_size = 0.6, random_state=random_state)
    print('training set:', X_train.shape, Y_train.shape)
    print(X_other.shape, Y_other.shape)
    X_val, X_test, Y_val, Y_test = train_test_split(X_other, Y_other, train_size = 0.5, random_state=random_state)
    print('validation set:', X_val.shape, Y_val.shape)
    print('test set:', X_test.shape, Y_test.shape)
    
    # OneHotEncoder for catagorical feature; fitted to train set only
    enc = OneHotEncoder(sparse=False, handle_unknown='ignore')
    # enc = OneHotEncoder(sparse=False)
    enc.fit(X_train[onehot_features])
    print('categories:', enc.categories)
    print('feature names:', enc.get_feature_names(onehot_features))
    
    X_train_onehot = enc.fit_transform(X_train[onehot_features])
    print('X_train_onehot transformed')
    print(X_train_onehot)
    # OneHotEncoder transform applied to validation and test sets
    X_val_onehot = enc.transform(X_val[onehot_features])
    print('X_val_onehot transformed')
    print(X_val_onehot)
    X_test_onehot = enc.transform(X_test[onehot_features])
    print('X_test_onehot transformed')
    print(X_test_onehot)
    
    # Initialize ordinal encoder
    # ord_enc = OrdinalEncoder(categories = ordinal_catagories)
    # Ordinal Encoder fit to training set only
    # ord_enc.fit(X_train[ordinal_features])
    
    # X_train_ord = ord_enc.transform(X_train[ordinal_features])
    # print('transformed Ordinal train features:')
    # print(X_train_ord)
    # Ordinal transform on validation and test sets.
    # X_val_ord = ord_enc.transform(X_val[ordinal_features])
    # print('transformed Ordinal validation features:')
    # print(X_val_ord)
    # X_test_ord = ord_enc.transform(X_test[ordinal_features])
    # print('transformed Ordinal test features:')
    # print(X_test_ord)
    # print('feature names:',enc.get_feature_names(ordinal_features))
    
    # Initialize MinMax Scaler
    min_max = MinMaxScaler()
    # Fit MinMax scaler on training set only
    min_max.fit(X_train[min_max_features])
      
    X_train_min_max = min_max.transform(X_train[min_max_features])
    print('transformed MinMax train features:')
    print(X_train_min_max)
    X_val_min_max = min_max.transform(X_val[min_max_features])
    print('transformed MinMax validation features:')
    print(X_val_min_max) 
    X_test_min_max = min_max.transform(X_test[min_max_features]) 
    print('transformed MinMax test features:')
    print(X_test_min_max) 
    
    # Initialize Standard Scaler
    standard_scl = StandardScaler()
    # Fit Standard scaler on training set only
    standard_scl.fit(X_train[standard_features])
      
    X_train_standard = standard_scl.transform(X_train[standard_features])
    print('transformed Standard train features:')
    print(X_train_standard)
    X_val_standard = standard_scl.transform(X_val[standard_features])
    print('transformed Standard validation features:')
    print(X_val_standard) 
    X_test_standard = standard_scl.transform(X_test[standard_features]) 
    print('transformed Standard test features:')
    print(X_test_standard)
```

```python
# TESTING CELL
# This cell is using one random state for testing pipeline and model
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression



train_MSE = np.zeros(len(random_states))
val_MSE = np.zeros(len(random_states))
test_MSE = np.zeros(len(random_states))
train_r2 = np.zeros(len(random_states))
val_r2 = np.zeros(len(random_states))
test_r2 = np.zeros(len(random_states))
state = np.zeros(len(random_states))

# Loop through pipeline with multiple random states.
for i in range(len(random_states)):  
    state[i] = i + 1
    # 60% training, 20% validation, 20% split test
    X_train, X_other, Y_train, Y_other = train_test_split(x, y, train_size = 0.6, 
                                                          random_state=random_states[i])
    
    print('training set:', X_train.shape, Y_train.shape)
    print(X_other.shape, Y_other.shape)
    X_val, X_test, Y_val, Y_test = train_test_split(X_other, Y_other, train_size = 0.5, 
                                                    random_state=random_states[i])
    
    print('validation set:', X_val.shape, Y_val.shape)
    print('test set:', X_test.shape, Y_test.shape)
    
    # Encoders for preprocessing
    preprocessor = ColumnTransformer(
    transformers=[
        #('ord', OrdinalEncoder(categories = ordinal_catagories), ordinal_features),
        ('onehot', OneHotEncoder(sparse=False, handle_unknown='ignore'), onehot_features),
        ('min_max', MinMaxScaler(), min_max_features),
        ('standard_scl', StandardScaler(), standard_features)])

    prep = Pipeline(steps=[('preprocessor', preprocessor)])  
                                                       

    X_train_prep = prep.fit_transform(X_train)
    X_val_prep = prep.transform(X_val)
    X_test_prep = prep.transform(X_test)

    print(X_train.shape)
    print(X_train_prep.shape)
    print(X_train_prep)
    
    # let's use only the first feature
    linreg = LinearRegression(fit_intercept=True)
    # linreg.fit(X_train_prep[:,:1], Y_train)
    linreg.fit(X_train_prep, Y_train)
    print('intercept:',linreg.intercept_)
    print('theta:',linreg.coef_)
    
    # R2 Score metric
    train_r2[i] = r2_score(Y_train, linreg.predict(X_train_prep))
    val_r2[i] = r2_score(Y_val, linreg.predict(X_val_prep))
    test_r2[i] = r2_score(Y_test, linreg.predict(X_test_prep))
    print('train R2:',train_r2[i])
    print('val R2:',val_r2[i])
    print('test R2:',test_r2[i])
       
```

```python
# R2 scores for linear regression 
plt.title('R2 Scores for Linear Regression Per Random State')
plt.plot(state,train_r2,label='train score')
plt.plot(state,val_r2,label='val score')
plt.plot(state,test_r2,label='test score')
plt.xlabel('random state')
plt.ylabel('R2')
plt.legend()
plt.tight_layout()
plt.savefig('../figures/train_val_test_R2_linear.png',dpi=300)
plt.show()
```

## Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

# Random state used in splitting process.
random_states = [789]

# List of tree depths
max_depth = [1, 2, 3, 5, 10]

train_MSE_rfor = np.zeros(len(max_depth))
val_MSE_rfor = np.zeros(len(max_depth))
test_MSE_rfor = np.zeros(len(max_depth))
train_r2_rfor = np.zeros(len(max_depth))
val_r2_rfor = np.zeros(len(max_depth))
test_r2_rfor = np.zeros(len(max_depth))
depth_num = np.zeros(len(max_depth))


# Loop through pipeline with multiple tree depths.
for l in range(len(max_depth)):  
    depth_num[l] = l + 1
    # 60% training, 20% validation, 20% split test
    X_train, X_other, Y_train, Y_other = train_test_split(x, y, train_size = 0.6, 
                                                          random_state=789)
    print('training set:', X_train.shape, Y_train.shape)
    print(X_other.shape, Y_other.shape)
    X_val, X_test, Y_val, Y_test = train_test_split(X_other, Y_other, train_size = 0.5, 
                                                    random_state=789)
    print('validation set:', X_val.shape, Y_val.shape)
    print('test set:', X_test.shape, Y_test.shape)
    
    # Encoders for preprocessing
    preprocessor = ColumnTransformer(
    transformers=[
        #('ord', OrdinalEncoder(categories = ordinal_catagories), ordinal_features),
        ('onehot', OneHotEncoder(sparse=False, handle_unknown='ignore'), onehot_features),
        ('min_max', MinMaxScaler(), min_max_features),
        ('standard_scl', StandardScaler(), standard_features)])

    prep = Pipeline(steps=[('preprocessor', preprocessor)])  
                                                       

    X_train_prep = prep.fit_transform(X_train)
    X_val_prep = prep.transform(X_val)
    X_test_prep = prep.transform(X_test)

    print(X_train.shape)
    print(X_train_prep.shape)
    print(X_train_prep)
    
    # Initialize Random Forest Regressor
    rand_for = RandomForestRegressor(n_estimators=100,max_depth=max_depth[l])
    
    # Fit Random Forest Regressor to training set
    rand_for.fit(X_train_prep, Y_train)
    
    # Predict Random Forest Regressor for training, validation and test sets
    y_new_train = rand_for.predict(X_train_prep)
    y_new_val = rand_for.predict(X_val_prep)
    y_new_test = rand_for.predict(X_test_prep)
    
    # Mean Squared Error metric
    # train_MSE = mean_squared_error(Y_train,linreg.predict(X_train_prep[:,:1]))
    train_MSE_rfor[l] = mean_squared_error(Y_train, y_new_train)
    # val_MSE = mean_squared_error(Y_val,linreg.predict(X_val_prep[:,:1]))
    val_MSE_rfor[l] = mean_squared_error(Y_val, y_new_val)
    test_MSE_rfor[l] = mean_squared_error(Y_val, y_new_test)
    print('train MSE:',train_MSE_rfor[l])
    print('val MSE:',val_MSE_rfor[l])
    print('test MSE:',test_MSE_rfor[l])
    
    # R2 Score metric
    train_r2_rfor[l] = r2_score(Y_train, y_new_train)
    val_r2_rfor[l] = r2_score(Y_val, y_new_val)
    test_r2_rfor[l] = r2_score(Y_test, y_new_test)
    print('train R2:',train_r2_rfor[l])
    print('val R2:',val_r2_rfor[l])
    print('test R2:',test_r2_rfor[l])
    
    
    
```
